```python
from google.cloud import storage

def list_bucket_objects(bucket_name):
    storage_client = storage.Client()
    bucket = storage_client.get_bucket(bucket_name)
    blobs = list(bucket.list_blobs())
    
    for blob in blobs:
        print(f"Object name: {blob.name}")

if __name__ == "__main__":
    bucket_name = "meta_ads_creatives"
    list_bucket_objects(bucket_name)
```

```python
def list_public_urls(bucket_name):
    storage_client = storage.Client()
    bucket = storage_client.get_bucket(bucket_name)
    blobs = list(bucket.list_blobs())

    for blob in blobs:
        public_url = f"https://storage.googleapis.com/{bucket_name}/{blob.name}"
        print(f"Public URL for {blob.name}: {public_url}")

if __name__ == "__main__":
    bucket_name = "meta_ads_creatives"
    list_public_urls(bucket_name)
```

```python
import io
from google.cloud import storage

def upload_image_file(bucket_name, source_file_name, destination_blob_name):
  """Uploads an image file to the bucket."""
  storage_client = storage.Client()
  bucket = storage_client.get_bucket(bucket_name)
  blob = bucket.blob(destination_blob_name)

  with io.open(source_file_name, 'rb') as f:
    blob.upload_from_file(f)

  print('File {} uploaded to {}.'.format(source_file_name, destination_blob_name))

# Example usage:

bucket_name = 'meta_ads_creatives'
source_file_name = 'C:\\Users\\jiri.pillar\\Developer\\python\\meta_ads_crawler\\downloaded_images\\1731653497124334\\135626122974379.jpg'
destination_blob_name = '135626122974379.jpg'

upload_image_file(bucket_name, source_file_name, destination_blob_name)
```
