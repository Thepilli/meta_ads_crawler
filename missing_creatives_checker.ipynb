```python
import os
import json
# Path to the folder containing JSON files.
folder_path = 'C:\\Users\\jiri.pillar\\Developer\\python\\meta_ads_crawler\\meta_data_ids'

# Initialize an empty list to store the extracted IDs.
all_ids = []

# Iterate through the JSON files in the folder.
for filename in os.listdir(folder_path):
    if filename.endswith('.json'):
        file_path = os.path.join(folder_path, filename)
        
        # Open the JSON file and load its content.
        with open(file_path, 'r') as file:
            data = json.load(file)
        
        # Extract the 'id' values and append them to the 'all_ids' list.
        for item in data:
            if 'id' in item:
                all_ids.append(item['id'])
print("Total number of IDs:", len(all_ids))
```

```python
from google.cloud import storage

# Create a client with a reference to the bucket
bucket_name = "meta_ads_creatives"
storage_client = storage.Client()
bucket = storage_client.get_bucket(bucket_name)
blobs = list(bucket.list_blobs())
# Extract the file IDs without the ".jpg" extension and store them in a list.
file_ids = [blob.name.split('.')[0] for blob in blobs]

print("Total number of IDs:", len(file_ids))
```

```python
# Convert the lists to sets for efficient comparison
json_ids_set = set(all_ids)
file_ids_set = set(file_ids)

# Find the missing IDs that are in json_ids but not in file_ids
missing_ids = json_ids_set - file_ids_set

# Convert the missing IDs back to a list
missing_ids_list = list(missing_ids)

# Print the missing IDs
print("Total number of missing IDs:",len(missing_ids_list))
print("Missing IDs:", missing_ids_list)
```
