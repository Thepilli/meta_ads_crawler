## get and save image urls from github


```python
import requests
import csv


def get_raw_file_urls(username, repo, path):
    api_url = f"https://api.github.com/repos/{username}/{repo}/contents/{path}"
    response = requests.get(api_url)

    if response.status_code == 200:
        files = response.json()
        raw_file_urls = [file['download_url'] for file in files]
        return raw_file_urls
    else:
        print(f"Failed to fetch files. Status code: {response.status_code}")
        return None
    
def save_urls_to_csv(file_urls, output_file):
    with open(output_file, 'w', newline='') as csvfile:
        url_writer = csv.writer(csvfile)
        url_writer.writerow(['image_url'])
        for url in file_urls:
            url_writer.writerow([url])

# Replace these values with your GitHub username, repository name, and folder path
username = "Thepilli"
repo = "meta_ads_political"
folder_path = "page_name_images"
output_file = "page_name_images.csv"

raw_file_urls = get_raw_file_urls(username, repo, folder_path)

if raw_file_urls:
    save_urls_to_csv(raw_file_urls, output_file)
    print(f"Raw File URLs saved to {output_file}")
```
